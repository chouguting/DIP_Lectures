```python
from skimage import io
import matplotlib.pyplot as plt


img = io.imread('result.jpg')
io.imshow(img)
```

```python
y=img-128
y[y<0]=0
io.imshow(y)
```

```python
#所有值除以2
y = img/2
io.imshow(y,cmap='gray')
```

```python
y = img*2
#對於大於255的 設成255
y[y<100]=150

io.imshow(y)
```

```python
y = 255-img
io.imshow(y,cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
#x軸
a = np.array([0, 100, 150, 255])
#y軸
b = np.array([40,100, 220, 255])
#定義一條線 (0~256),自動生成圖
lin = np.interp(range(256), a, b)

#0~256的每個點
x = range(256)

#秀出圖
f = plt.figure()
f.show(plt.plot(x,lin))
```

```python
#gamma處理
import  skimage.exposure as ex
gamma2Pic=ex.adjust_gamma(img,2)
io.imshow(gamma2Pic,cmap="gray")
```

```python
#生成直方圖
f= plt.figure()
#bins:切割數量
#flatten:把二維陣列轉成一維
f.show(plt.hist(gamma2Pic.flatten(), bins =256))
```

```python
#直方圖等化
cal = ex.equalize_hist(gamma2Pic)
io.imshow(cal)
f= plt.figure()
f.show(plt.hist(cal.flatten(), bins =256))
```
